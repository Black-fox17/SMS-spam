```python
df = pd.read_csv("/kaggle/input/sms-spam-collection-dataset/spam.csv",encoding = "latin-1")
```

```python
df.head()
```

```python
df.info()
```

```python
def unique(column):
    unique = df[column].unique()
    return unique
```

```python
null_columns = ["Unnamed: 2","Unnamed: 3","Unnamed: 4"]
for null_column in null_columns:
    print(null_column)
    print(unique(null_column))
    print()
```

```python
new_df = df.drop(null_columns,axis = 1)
new_df.head()
```

```python
new_df.columns = ["spam","text"]
new_df["spam"] = new_df["spam"].map({"ham":0,"spam":1})
new_df.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(new_df.text,new_df["spam"])
x_train.shape
```

```python
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x_train_vec = vectorizer.fit_transform(x_train)
x_train_np = x_train_vec.toarray()
x_train_np
```

```python
x_test_vec = vectorizer.transform(x_test)
x_test_np = x_test_vec.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_np,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
y_pred = model.predict(x_test_np)
print(classification_report(y_test,y_pred))
```

```python
print("Accuracy score: ",accuracy_score(y_test,y_pred))
```

```python
def predict(text):
    transform = vectorizer.transform([text])
    trans_arr = transform.toarray()
    pred = model.predict(trans_arr)
    if 0 in pred:
        return "Not spam"
    return "Spam"
predict("Hey how have you been jota")
```

```python
predict("Claim 30 dollars from our website")
```
